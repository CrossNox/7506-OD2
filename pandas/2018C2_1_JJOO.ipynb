 Olímpicos de la Juventud por parte del público, sus organizadores realizan distintos análisis para planificar las jornadas finales del certamen. Es por ello que cuentan con información en los siguientes archivos csv: 
 
eventos.csv (id_evento, fecha, id_locacion, nombre_evento, id_categoria_deportiva, cantidad_espectadores) 

locacion.csv  (id_locacion, nombre, capacidad, capacidad_extendida, sede, latitud, longitud) 

categorias_deportivas.csv (id_categoria_deportiva, nombre, año_de_adopcion) 

El primer archivo cuenta con información de los eventos, indicando la fecha (en formato “YYYY-mm-dd hh:mm:ss”), el lugar donde ocurrió (id_locacion) y la cantidad de espectadores que asistieron. Además se aporta información sobre la categoría deportiva a la cual pertenece el evento. 

Por  otro  lado se tiene  información sobre las distintas locaciones en la sedes  del certamen en las que ocurrieron  los eventos. Contamos con información de su capacidad total de espectadores así como de su capacidad extendida (cuantos asientos extras se pueden brindar sobre la capacidad de la locación). 

Se desea obtener: 

a) Nombre de la sede que acumuló la mayor cantidad de espectadores en eventos durante el certamen del 14 al 15 de octubre inclusive. Esto es de vital importancia para distribuir el merchandising oficial del evento, para las fechas finales. (7 pts) 

b) Nombre del evento y nombre de la categoría deportiva de aquellos eventos cuya cantidad de espectado
res superó la capacidad de la locación, más allá de la capacidad extendida. Esto es de vital importancia para detectar problemas de seguridad o si es necesario realizar algún cambio de locación. (8 pts)

```python
# Set-up y vista rápida de los dos sets de datos sintéticos
import pandas as pd

df_eventos = pd.read_csv('../data/2018C2_1_eventos.csv')
display(df_eventos)

df_categorias_deportivas = pd.read_csv('../data/2018C2_1_categorias_deportivas.csv')
display(df_categorias_deportivas)

df_locacion = pd.read_csv('../data/2018C2_1_locacion.csv')
display(df_locacion)
```

## Punto A

```python
df_eventos = df_eventos[['id_loc','fecha','cant']]
df_eventos['fecha'] = pd.to_datetime(df_eventos['fecha'])
df_eventos = df_eventos.loc[(df_eventos['fecha'].dt.day == 14)| (df_eventos['fecha'].dt.day == 15)]

df_eventos.head()
```

```python
df_eventos = df_eventos[['id_loc','cant']]
df_locacion = df_locacion[['id_loc','sede']]
df_eventosxsede = df_eventos.merge(df_locacion,on='id_loc')

df_eventosxsede.head()
```

```python
df_eventosxsede = df_eventosxsede[['sede','cant']]
df_eventosxsede = df_eventosxsede.groupby('sede').agg({'cant':'sum'})

df_eventosxsede.head()
```

```python
df_eventosxsede['cant'].nlargest(1)
```

```python
df_locacion['cap+ext'] = df_locacion['capacidad'] + df_locacion['capacidad_extendida']
df_locacion = df_locacion[['id_loc','cap+ext']]

df_locacion.head()
```

```python
df_eventosconloc = df_eventos[['id_loc','nombre_ev','id_cat','cant']].merge(df_locacion, on='id_loc')

df_eventosconloc.head()
```

```python
df_eventosconloc = df_eventosconloc.loc[df_eventosconloc['cant'] > df_eventosconloc['cap+ext']]

df_eventosconloc.head()
```
